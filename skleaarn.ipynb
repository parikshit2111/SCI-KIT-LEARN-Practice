# 0.An end-to-end sklearn workflow

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
heart_disease=pd.read_csv("heart-disease.csv")
heart_disease
```

```python
#create X (features matrix) removing target column and storing the dataset in X
X = heart_disease.drop("target",axis=1)

# create y (lables) string target values in y
y=heart_disease["target"]
```

```python
heart_disease
```

```python
y
```

```python
# 2. choose the right model and hyperparameters

from sklearn.ensemble import RandomForestClassifier

clf=RandomForestClassifier()

# keep the default hyperparameters
clf.get_params()
```

```python
# 3.fit the model to the data

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

# 0.2 means 20%of the data is left for testing and 80% is used for training
```

```python
clf.fit(  X_train,y_train  )
```

```python
#make prediction

# y_label=clf.predict(np.array([0,2,3,4]))

# above will not work because machine did not learn to the shape of np array 

# thus we will use the testing data as the shape will be same as the training data.
```

```python
X_train
```

```python
X_test
```

```python
# y_preds=y predict
y_preds=clf.predict(X_test)
y_preds
```

```python
y_test
```

```python
#4. evaluate the model on training data and test data

clf.score(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score

print(classification_report(y_test,y_preds))
```

```python
confusion_matrix(y_test,y_preds)
```

```python
accuracy_score(y_test,y_preds)
```

```python
# 5.improve a model 
#try different amount of n_estimators
np.random.seed(42)
for i in range (10,100,10):
    print(f"tring model with {i} estimators")
    clf=RandomForestClassifier(n_estimators=i).fit(X_train,y_train)
    print(f"model accuracy on test set :{clf.score(X_test,y_test)*100:2f}%")
    print("")
```

```python
#6 save a model and load it
import pickle

pickle.dump(clf,open("random_forest_model_1.pkl","wb"))
#wb=write binary
```

```python
loaded_model=pickle.load(open("random_forest_model_1.pkl" ,"rb"))
#rb=read binaries
loaded_model.score(X_test,y_test)
```
